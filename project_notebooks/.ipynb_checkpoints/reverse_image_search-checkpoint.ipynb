```python
import numpy as np
import matplotlib.pyplot as plt

from tqdm import tqdm
import cv2
import os 

import keras
from keras.preprocessing import image
from keras.applications.imagenet_utils import decode_predictions, preprocess_input
from keras.models import Model
```

```python
#dowloading data for feature extraction
model = keras.applications.VGG16(weights='imagenet', include_top=True)
```

```python
model.summary()
```

```python
def convert_image(image):
    ''' Changes BGR to RGB'''
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

def show_image(image):
    plt.imshow(convert_image(image))
    plt.axis("off")
    plt.show()
```

```python
#loading numpy arrays 
X = np.load('X_data.npy')
labels = np.load('labels_data.npy')
```

```python
#feature extraction
feat_extractor = Model(inputs=model.input, outputs=model.get_layer("fc2").output)
#feat_extractor.summary()
```

```python
#using pre-trained "extractor"
feat = feat_extractor.predict(X[:1])
```

```python
# plot of activations in last layer "fc2"
plt.figure(figsize=(16,4))
plt.plot(feat[0])
```

```python
#extracting features for every image
features = []
for x in tqdm(X):
    features.append(feat_extractor.predict(np.expand_dims(x, 0))[0])
```

```python
#pca for dimensionality reduction
from sklearn.decomposition import PCA

features = np.array(features)
#we keep 300 top principal components
pca = PCA(n_components=300)
pca.fit(features)
```

```python
pca_features = pca.transform(features)
```

```python
import random 

rand_index = random.randint(0, len(X))
print(labels[rand_index])
show_image(X[rand_index])
```

```python
from scipy.spatial import distance

#"reverse search" using similiarty via cosine distance returns distances
similar_idx = [ distance.cosine(pca_features[rand_index], feat) for feat in pca_features ]
#5 closest indexes (from 1 because image for which we want to find similar images is also included)
idx_closest = sorted(range(len(similar_idx)), key=lambda k: similar_idx[k])[1:6]
```

```python
for idx in idx_closest:
    show_image(X[idx])
```

```python
def get_image(url):
    ''' 
    downloads from local file and prepares to use in reverse search, 
    in case of use in show_image() add [0, :, :, :] to delete batch dimension
    '''
    image =  cv2.resize(cv2.imread(os.path.join(url)), (224, 224))
    return np.array(image).reshape(-1, 224, 224, 3)


def get_closest_to_image(image, N=5):
    """
    input is straight numpy array (not features array)
    returns N closest images indexes
    """
    new_pca_features = pca.transform(feat_extractor.predict(image))[0]
    distances = [ distance.cosine(new_pca_features, feat) for feat in pca_features ]
    return sorted(range(len(distances)), key=lambda k: distances[k])[0:5] 
    
```

```python
#showing random image and closest to it ;]
n = random.randint(0, 1800)
print(labels[n])
show_image(X[n])
```

```python
for idx in get_closest_to_image(np.expand_dims(X[n], 0))[1:]:
    print(labels[idx])
    show_image(X[idx])
```
