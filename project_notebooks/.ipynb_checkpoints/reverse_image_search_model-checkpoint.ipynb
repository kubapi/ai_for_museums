## Method - PCA (Principal Component Analysis)

```python
def get_image(url):
    ''' 
    downloads from local file and prepares to use in reverse search, 
    in case of use in show_image() add [0, :, :, :] to delete batch dimension
    '''
    image =  cv2.resize(cv2.imread(os.path.join(url)), (224, 224))
    return np.array(image).reshape(-1, 224, 224, 3)


def PCA_get_closest_to_image(image, N=5):
    """
    input is straight numpy array (not features array)
    returns N closest images indexes
    """
    new_pca_features = pca.transform(feat_extractor.predict(image))[0]
    distances = [ distance.cosine(new_pca_features, feat) for feat in pca_features ]
    return sorted(range(len(distances)), key=lambda k: distances[k])[0:N] 

def LSA_get_closest_to_image(image, N=5):
    """
    input is straight numpy array (not features array)
    returns N closest images indexes
    """
    new_lsa_features = lsa.transform(feat_extractor.predict(image))[0]
    distances = [ distance.cosine(new_lsa_features, feat) for feat in lsa_features ]
    return sorted(range(len(distances)), key=lambda k: distances[k])[0:N] 

def compare_view(image, N=5):
    pass
    
```

```python
#we applay PCA and then we keep 300 top principal components
pca = PCA(n_components=300)
pca.fit(features)
pca_features = pca.transform(features)
```

```python
#showing random image and closest to it ;]
n = random.randint(0, 1800)
show_image(X[n], labels[n])

print("\n")

for idx in PCA_get_closest_to_image(np.expand_dims(X[n], 0))[1:]:
    show_image(X[idx], title = str(labels[idx]))
```

## Method for visualization - t-SNE (t-distributed Stochastic Neighbor Embedding)

```python
tsne = TSNE(n_components=2)
#using PCA top components (t-SNE doesn't like big number of featurers)
tsne_features = tsne.fit_transform(pca_features)
```

```python
tx, ty = tsne_features[:,0], tsne_features[:,1]
tx = (tx-np.min(tx)) / (np.max(tx) - np.min(tx))
ty = (ty-np.min(ty)) / (np.max(ty) - np.min(ty))
```

```python
# from PIL import Image
width = 4000
height = 3000
max_dim = 100

full_image = Image.new('RGBA', (width, height))
iter = 0
for x, y in zip(tx, ty):
    tile = Image.open(get_image_path_by_label(labels[iter]))
    rs = max(1, tile.width/max_dim, tile.height/max_dim)
    tile = tile.resize((int(tile.width/rs), int(tile.height/rs)), Image.ANTIALIAS)
    full_image.paste(tile, (int((width-max_dim)*x), int((height-max_dim)*y)), mask=tile.convert('RGBA'))
    iter += 1
```

```python
full_image.save("tSNE.png")
```
